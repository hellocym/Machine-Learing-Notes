```python
# 预处理
def pre_process(img):
    # 转化为列表
    lst = img.split('\n')
    
    # 转化为由0和1组成的矩阵
    width = len(lst[0])
    height = len(lst)
    img = [[0 if lst[i][j] == '.' else 1 for j in range(width)] for i in range(height)]
    return img
```

```python
# 高斯函数
def G(x, y ,sigma):
    return 1/(2 * math.pi * sigma**2) * math.e**(-(x**2 + y**2) / (2 * sigma**2))
```

```python
# 生成卷积核
def CoreGenerate(sigma, r):
    a = 1 + 2 * r
    core = [[G(x, y, sigma) for x in range(-r, r + 1)] for y in range(r, -r - 1, -1)]
    ave = sum(map(sum, core))
    core = [[x / ave for x in row] for row in core]
    return core
```

```python
# 高斯滤波
def GaussianBlur(img, sigma, r):
    core = CoreGenerate(sigma, r)
    a = 1 + r * 2
    width = len(img[0])
    height = len(img)
    result = [[0] * width for _ in range(height)]
    for i in range(r, height - r):
        for j in range(r, width - r):
            img_slice = [[img[x][y] for y in range(j-r, j+r+1)] for x in range(i-r, i+r+1)]
            result[i][j] = round(sum([img_slice[i][j] * core[i][j] for i in range(a) for j in range(a)]), 0)
    return result
```

```python
# 分割数字
def ImgSplit(img):
    width = len(img[0])
    height = len(img)
    x = 0
    result = []
    # 找第1个数字的左边界
    while [img[y][x] for y in range(height)].count(0) == height:
        x += 1
    left_border = x
    
    while x < width:
        # 找第cnt个间隔的左边界
        while [img[y][x] for y in range(height)].count(0) != height:
            x += 1
        right_border = x
        
        # 找第cnt+1个数字的左边界
        while x < width and [img[y][x] for y in range(height)].count(0) == height:
            x += 1
        
        tmp = [[img[y][x] for x in range(left_border, right_border + 1)] for y in range(height)]
        result.append(tmp)
        left_border = x
    
    return result
```
